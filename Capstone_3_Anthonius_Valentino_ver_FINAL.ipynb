# **A. Business Problem Understanding**

## A.1 Context

Pada industri perhotelan revenue management adalah memastikan kamar yang tepat tersedia untuk tamu yang tepat dan dengan harga yang tepat pada saat yang tepat melalui distribution channel yang tepat pula. Hotel memiliki fixed inventory artinya jumlah kamar yang tersedia tidak dapat dikurangi atau ditambah. Kamar ini dianggap sebagai perishable product karena sebuah kamar yang tidak terjual atau tidak ditempati akan menjadi opportunity cost. Untuk menanggulangi hal ini hotel menerima booking di muka sebagai kontrak antara calon tamu dengan hotel.

Kontrak dari booking di muka memungkinkan tamu untuk menggunakan layanan hotel dengan harga tetap di masa depan, terkadang kontrak ini juga memiliki opsi untuk dibatalkan sebelum kedatangan. Meskipun booking dapat digunakan sebagai faktor untuk forecasting performa hotel, pembatalan booking menjadi resiko bagi hotel itu sendiri. Hotel harus memastikan kamar tersedia bagi calon tamu yang menepati kontrak tersebut, namun pada saat yang bersamaan harus menanggung opportunity cost dari kapasitas yang tidak terisi ketika tamu tidak menepati booking atau tidak datang. 

## A.2 Problem Statement

**Analisis Efisiensi Operasional Hotel dengan Menggunakan 6 Metrik Utama**

Dalam mengukur efisiensi operasional sebuah hotel, terdapat enam indikator atau metrik utama yang digunakan, yaitu Cost per Occupied Room (CPOR), Revenue per Available Room (RevPAR), Cost per Available Room (CostPAR), Gross Operating Profit per Available Room (GOPPAR), Labor per Available Room (LPAR), dan Guest Acquisition Costs (GAC). Berikut ini adalah penjelasan dan contoh perhitungan menggunakan data hipotetikal dari Hotel A:

**1. Cost per Occupied Room (CPOR)**

CPOR mengukur rata-rata biaya yang dikeluarkan hotel untuk setiap kamar yang ditempati tamu. Semakin rendah nilai CPOR, semakin tinggi potensi profitabilitas. 

Rumus perhitungannya adalah:
- CPOR = Total Operating Expenses / Total Number of Occupied Rooms
- Contoh perhitungan:
- Total Operating Expenses: $500,000
- Total Number of Occupied Rooms: 70
- CPOR = $500,000 / 70 = $7,142.86
- Nilai CPOR sebesar $7,142.86 menunjukkan bahwa setiap kamar yang terisi dikenakan biaya operasional sebesar itu. Jika angka ini besar, ini menunjukkan adanya ketidakefisienan dalam manajemen biaya operasional.

**2. Revenue per Available Room (RevPAR)**

RevPAR menunjukkan jumlah pendapatan yang dihasilkan dari kamar yang tersedia sebagai inventori, terlepas dari apakah kamar tersebut terisi atau tidak. 

Rumus perhitungannya adalah:
- RevPAR = Total Room Revenue / Total Number of Available Rooms
- Contoh perhitungan:
- Total Room Revenue: $1,200,000
- Total Number of Available Rooms: 100 (365 days)
- RevPAR = $1,200,000 / (100 x 365) = $32.88
- RevPAR sebesar $32.88 menunjukkan jumlah pendapatan dari kamar yang tersedia. Semakin tinggi nilai RevPAR, semakin baik performa pendapatan hotel.

**3. Cost per Available Room (CostPAR)**

CostPAR mengukur rata-rata biaya yang dikeluarkan untuk seluruh kamar yang tersedia, terlepas dari apakah kamar tersebut terisi atau tidak. 

Rumus perhitungannya adalah:
- CostPAR = Total Costs / (Total Number of Available Rooms x Days)
- Contoh perhitungan:
- Total Costs: $500,000
- Total Number of Available Rooms: 100 (365 days)
- CostPAR = $500,000 / (100 x 365) = $13.70
- CostPAR sebesar $13.70 memberikan gambaran tentang struktur biaya hotel. Jika angka ini mendekati atau melebihi RevPAR, ini mengindikasikan bahwa biaya mungkin mengurangi profitabilitas.

**4. Gross Operating Profit per Available Room (GOPPAR)**

GOPPAR mengukur laba operasi bruto dari kamar yang tersedia dan memberikan gambaran lebih baik tentang performa finansial hotel. 

Rumus perhitungannya adalah:
GOPPAR = Gross Operating Profit (GOP) / Total Number of Available Rooms
Contoh perhitungan:
Gross Operating Profit (GOP): $700,000
Total Number of Available Rooms: 100 (365 days)
GOPPAR = $700,000 / (100 x 365) = $19.18
GOPPAR sebesar $19.18 menunjukkan nilai laba kotor per kamar tersedia.

**5. Labor per Available Room (LPAR)**

LPAR mengukur efisiensi sumber daya tenaga kerja relatif terhadap jumlah kamar yang tersedia.

Rumus perhitungannya adalah:
LPAR = Total Labor Costs / (Total Number of Available Rooms x Days)
Contoh perhitungan:
Total Labor Costs: $200,000
Total Number of Available Rooms: 100 (365 days)
LPAR = $200,000 / (100 x 365) = $5.48
LPAR sebesar $5.48 menunjukkan bagaimana biaya tenaga kerja didistribusikan di seluruh kamar yang tersedia.

**6. Guest Acquisition Costs (GAC)**

GAC mengukur biaya rata-rata untuk mengakuisisi tamu baru, 
dan dihitung dengan rumus:

GAC = (Guest Acquisition Costs / Total Room Revenue) x 100
Contoh perhitungan:
Guest Acquisition Costs: $150,000
Total Room Revenue: $1,200,000
GAC = ($150,000 / $1,200,000) x 100 = 12.5%
GAC sebesar 12.5% menunjukkan bahwa hotel menghabiskan 12.5% dari pendapatan kamarnya untuk mengakuisisi tamu baru. GAC yang tinggi dapat mengurangi profitabilitas, sehingga penting untuk memastikan strategi pemasaran dan distribusi yang digunakan lebih efektif.


Dampak Pembatalan Booking pada Metrik Operasional
Pembatalan booking yang signifikan selama peak seasons akan mempengaruhi keenam metrik ini. Misalnya:
1. CPOR akan meningkat jika kamar yang terisi menurun, misalnya menjadi 50 kamar, maka CPOR = $500,000 / 50 = $10,000.
2. RevPAR akan menurun jika pendapatan kamar turun menjadi $900,000, maka RevPAR = $900,000 / (100 x 365) = $24.66.
3. CostPAR mungkin menurun jika total biaya dikurangi menjadi $400,000, maka CostPAR = $400,000 / (100 x 365) = $10.96.
4. GOPPAR juga akan menurun jika GOP menurun menjadi $500,000, maka GOPPAR = $500,000 / (100 x 365) = $13.70.
5. LPAR akan menurun jika biaya tenaga kerja dikurangi menjadi $150,000, maka LPAR = $150,000 / (100 x 365) = $4.11.
6. GAC akan meningkat jika biaya akuisisi tamu meningkat menjadi $200,000, maka GAC = ($200,000 / $900,000) x 100 = 22.2%.

Pembatalan booking dapat mengakibatkan penurunan profitabilitas dan efisiensi operasional secara signifikan, sehingga penting untuk memonitor dan mengelola metrik-metrik ini secara efektif.

## A.5 Modelling

Dataset memiliki kolum is_canceled yang merupakan label dari baris data mengindikasikan tamu melakukan pembatalan booking atau tidak. Berdasarkan informasi tersebut model machine learning yang dibuat merupakan model supervised klasifikasi karena memiliki label. Beberapa model yang dapat digunakan pada kasus klasifikasi antara lain:

1. Logistic Regression

    Logistic regression adalah model sederhana namun efektif untuk membuat prediksi kemungkinan sebuah kejadian misalnya dalam kasus ini adalah pembatalan booking berdasarkan feature-feature dari calon tamu hotel. Model ini memiliki kelebihan bahwa hasil prediksi cenderung mudah diinterpretasikan sehingga dapat menjadi titik awal untuk melakukan analisis.

2. KNeighborsClassifier

    KNeighborsClassifier bekerja dengan membandingkan setiap bookingan dengan tetangga terdekatnya dan membuat prediksi berdasarkan outcome yang paling sering terjadi diantara kelompok ini. Model ini dapat menangkap pola yang kompleks yang berguna ketika keputusan calon tamu untuk membatalkan bookingan dipengaruhi dari banyak faktor.

3. Decision Tree

    Decision Tree membagi data menjadi cabang berdasarkan nilai fiturnya kemudian menghasilkan berbagai outcome yang menunjukkan apakah sebuah booking dibatalkan atau tidak. Decision tree dapat menunjukkan secara jelas faktor apa saja yang penting ketika memprediksi calon tamu akan melakukan pembatalan atau tidak.

4. Voting Classifer (Hard)

    Voting Classifier (Hard) menggabungkan beberapa model dan memprediksi outcome berdasarkan keputusan mayoritas. Pendekatan ini meningkatkan akurasi dengan memanfaatkan kelebihan dari beberapa model sehingga prediksi lebih reliabel.

5. Voting Classifier (Soft)

    Voting Classifier (Soft) menggabungkan prediksi dari beberapa model dengan menghitung rata-rata probabilitas yang diprediksi oleh setiap model dan memilih kelas dengan probabilitas rata-rata tertinggi. Pendekatan ini memungkinkan classifier mempertimbangkan tidak hanya hasil yang paling mungkin dari setiap model, tetapi juga tingkat kepercayaan dalam prediksi tersebut, sehingga menghasilkan prediksi yang lebih halus dan sering kali lebih akurat.

6. Stacking Classifier

    Stacking Classifier adalah metode ensemble yang menggabungkan beberapa model machine learning dengan melatih meta-model untuk membuat prediksi akhir. Setiap model dasar membuat prediksinya, dan meta-model menggunakan prediksi ini sebagai input untuk mempelajari cara menggabungkannya guna mencapai akurasi keseluruhan yang terbaik. 

7. Random Forest Classifier

    Random Forest Classifier adalah metode ensemble yang membangun beberapa Decision Tree dan menggabungkan prediksi mereka untuk meningkatkan akurasi dan mengurangi overfitting. Setiap pohon dalam hutan dilatih pada subset data dan fitur yang dipilih secara acak, dan prediksi akhir dibuat dengan mengambil suara mayoritas dari semua pohon. 

8. AdaBoost Classifier

    AdaBoost Classifier adalah teknik ensemble yang meningkatkan kinerja model-model lemah, biasanya Decision Tree, dengan memfokuskan pada instance yang salah diklasifikasikan oleh model sebelumnya. Metode ini bekerja dengan memberikan bobot lebih tinggi pada instance yang salah klasifikasi dan membangun rangkaian model yang bertujuan untuk memperbaiki kesalahan dari pendahulunya. 


9. GradientBoosting Classifier

    Gradient Boosting Classifier adalah metode ensemble yang membangun model secara berurutan, di mana setiap model baru memperbaiki kesalahan yang dibuat oleh model sebelumnya. Metode ini bekerja dengan meminimalkan fungsi kerugian dan secara bertahap meningkatkan akurasi prediksi.


10. XGB Classifier

    XGBClassifier adalah implementasi Gradient Boosting yang dioptimalkan untuk kecepatan dan kinerja. Ini menggabungkan teknik regularisasi canggih untuk mencegah overfitting, sehingga tetap tangguh bahkan dengan dataset yang kompleks. XGBoost sangat efisien dan skalabel, menjadikannya pilihan populer untuk menangani dataset besar dengan banyak fitur dan hubungan yang kompleks, seperti yang terlibat dalam memprediksi pembatalan hotel.

## A.6 Metrics Evaluation

**1. High Precision**

$$
\text{Precision} = \frac{\text{True Positives (TP)}}{\text{True Positives (TP)} + \text{False Positives (FP)}}
$$

Precision adalah proporsi dari prediksi positif yang secara aktual memiliki nilai positif. Presisi yang tinggi artinya ketika model memprediksi kasus positif maka kemungkinan besar secara aktual nilai nya memang positif. Hal ini menjadi penting pada skenario ketika false positif (memprediksi positif pada kasus yang secara aktual negatif) memiliki konsekuensi negatif. Pada kasus booking hotel presisi menjadi penting ketika cost untuk memberikan insentif yang berkaitan dengan upaya mengurangi pembatalan booking (seperto diskon, kebijakan pembatalan yang fleksibel, atau targeted marketing) tinggi sehingga manajemen hotel ingin memastikan bahwa insentif yang diberikan benar-benar kepada konsumen yang memiliki resiko pembatalan yang tinggi.

Scenario 1: Model memprioritaskan High Precision

**Cost per Occupied Room (CPOR)**
- Jika model dibuat dengan memprioritaskan precision, hotel akan secara akurat memprediksikan pembatalan dan mengindari preparation cost yang tidak perlu untuk booking yang tidak akan dibatalkan
- **Total Operating Expenses**: $480,000 (berkurang karena preparation cost yang lebih sedikit)
- **Total Number of Occupied Rooms**: 70
- **CPOR** = $480,0000 / 70 = **$6,857.14**

**Revenue per Avaliable Room (RevPAR)**
- Karena precision yang tinggi, hotel mungkin kehilangan sedikit revenue karena ada pembatalan yang tidak terprediksi (false negatives)
- **Total Room Revenue**: $1,150,000 (sedikit berkurang karena peluang yang terlewatkan)
- **Total Number of Available Rooms** : 100
- **RevPAR** = $1,150,000 / (100 x 365) = **$31.51**

**Guest Acquisition Costs (GAC)**
- Usaha pemasaran menjadi sangat ditargetkan ketika manajemen memprioritaskan high precision sehinggal pemborosan dapat dikurangi
- **Guest Acquisition Costs**: $145,000
- **GAC** = ($145,000/$1,150,000) x 100 = **12.61%**

**2. High Recall**

$$
\text{Recall} = \frac{\text{True Positives (TP)}}{\text{True Positives (TP)} + \text{False Negatives (FN)}}
$$

Recall (atau sensitivitas) adalah proporsi dari nilai yang secara aktual adalah positif dan diprediksi positif oleh model. Recall positif yang tinggi artinya model secara umum dapat mengidentifikasi semua kasus positif, meskipun memiliki nilai false positive. Hal ini menjadi penting ketika false negative atau kekeliruan dalam memprediksi kasus positif memiliki konsekuensi yang tinggi. Jika cost yang ditimbulkan pembatalan booking oleh calon tamu (misalnya, lost revenue, kekosongan kamar di saat-saat akhir) sangat tinggi maka recall yang tinggi harus diprioritaskan. Hal ini dilakukan untuk memastikan bahwa analis menangkap sebanyak mungkin potensi pembatalan, meskipun artinya harus memberikan pelayanan tambahan kepada pelanggan yang sebenarnya mungkin tidak akan membatalkan (false positive).

Scenario 2: Model memprioritaskan High Recall

**Cost per Occupied Room (CPOR)**
- Dengan fokus pada recall, hotel mempersiapkan lebih banyak kemungkinan terjadi pembatalan booking, termasuk kasus-kasus yang mungkin sebenarnya tidak terjadi. Hal ini akan meningkatkan potensi kenaikan cost
- **Total Operating Expenses**: $520,000 (meningkat karena preparation cost bertambah)
- **Total Number of Occupied Rooms**: 65 (karena lebih banyak pembatalan yang diantisipasi dan beberapa kamar yang tidak disiapkan)
- **CPOR** = $520,0000 / 65 = **$8,000**

**Revenue per Avaliable Room (RevPAR)**
- Meskipun terjadi peningkatan penanganan pembatalan, memprediksi kasus yag berpotensi melakukan pembatalan lebih banyak dapat membantu hotel untuk menangangi last minutes booking lebih baik
- **Total Room Revenue**: $1,210,000 (sedikit meningkat karena penanganan last minute booking yang lebih baik)
- **Total Number of Available Rooms** : 100
- **RevPAR** = $1,210,000 / (100 x 365) = **$33.15**

**Guest Acquisition Costs (GAC)**
- Usaha pemasaran dapat menjadi lebih luas untuk memastikan tidak ada tamu potensial yang melakukan pembatalan
- **Guest Acquisition Costs**: $160,000
- **GAC** = ($160,000/$1,210,000) x 100 = **13.22%**

**SUMMARY**
- **Skenario High Precision:**
CPOR: $6,857.14
RevPAR: $31.51
GAC: 12.61%

- **Skenario High Recall:**
CPOR: $8,000
RevPAR: $33.15
GAC: 13.22%

Penentuan Profitabilitas:

- Revenue per Available Room (RevPAR) lebih tinggi dalam skenario High Recall sebesar $33.15 dibandingkan dengan $31.51 dalam skenario High Precision.
- Cost per Occupied Room (CPOR) lebih rendah dalam skenario High Precision sebesar $6,857.14 dibandingkan dengan $8,000 dalam skenario High Recall.
- Guest Acquisition Costs (GAC) juga lebih rendah dalam skenario High Precision sebesar 12.61% dibandingkan dengan 13.22% dalam skenario High Recall.

**Profitabilitas Akhir:**

- High Recall menghasilkan pendapatan per kamar yang tersedia lebih tinggi ($33.15 vs. $31.51), tetapi juga menimbulkan biaya yang lebih tinggi (CPOR dan GAC).
High Precision menghasilkan biaya yang lebih rendah dan sedikit pendapatan yang lebih rendah.
Kesimpulan:

- High Precision kemungkinan akan lebih menguntungkan secara keseluruhan karena penghematan biaya (CPOR dan GAC yang lebih rendah) melebihi sedikit peningkatan pendapatan dalam skenario High Recall. Oleh karena itu, High Precision adalah skenario yang paling menguntungkan berdasarkan perhitungan yang disediakan.


**3. F1 Score, F2 Score dan F0.5 Score**

$$
\text{F1 Score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}
$$

F1 score adalah harmonic mean dari precision dan recall. F1 score dapat memberikan pandangan seimbang dari performa model, secara spesifik ketika ketika analis tidak dapat memutuskan dengan mudah apakah false positive atau false negative yang memiliki cost lebih tinggi. Hasil perhitungan performa ekonomi yang sudah dilakukan sebelumnya bahwa pembatalan booking dapat mengurangi revenue atau pendapatan (mempengaruhi RevPAR dan GOPPAR) dan meningkatkan cost (melalui naiknya CPOR dan GAC), dengan menggunakan F1 score dapat menyeimbangkan prioritas secara efektif. Optimasi dari model dapat dilakukan tanpa secara eksplisit memilih antara precision atau recall pada skenario dimana kedua tipe error adalah hal yang penting.

$$
\text{F2 Score} = 5 \times \frac{\text{Precision} \times \text{Recall}}{4 \times \text{Precision} + \text{Recall}}
$$

F2 score memberikan bobot lebih pada recall dibandingkan precision. Ini sangat berguna dalam skenario di mana false negatives (gagal memprediksi pembatalan) memiliki konsekuensi yang lebih tinggi dibandingkan false positives. Dalam konteks booking hotel, jika biaya yang ditimbulkan oleh gagal memprediksi pembatalan (seperti kehilangan revenue atau kamar kosong di saat-saat terakhir) jauh lebih tinggi daripada biaya yang terkait dengan prediksi pembatalan yang salah (false positive), maka F2 score adalah metrik yang lebih baik. Dengan memfokuskan model pada F2 score, hotel dapat memastikan bahwa sebanyak mungkin potensi pembatalan dapat diidentifikasi, meskipun itu berarti beberapa pembatalan yang diprediksi mungkin tidak terjadi, memastikan bahwa hotel lebih siap menghadapi potensi kehilangan revenue.

$$
\text{F0.5 Score} = 1.25 \times \frac{\text{Precision} \times \text{Recall}}{0.25 \times \text{Precision} + \text{Recall}}
$$

F0.5 score memberikan bobot lebih pada precision dibandingkan recall. Skor ini ideal dalam situasi di mana false positives memiliki biaya yang lebih tinggi dibandingkan false negatives. Dalam konteks pemesanan hotel, jika biaya terkait dengan salah memprediksi pembatalan (misalnya, memberikan diskon yang tidak perlu, menyesuaikan ketersediaan kamar) lebih besar daripada biaya kehilangan beberapa pembatalan yang tidak terprediksi, maka F0.5 score akan lebih tepat. Dengan memfokuskan model pada F0.5 score, hotel dapat meminimalkan jumlah false positives, memastikan bahwa sumber daya dan insentif hotel hanya diterapkan pada pemesanan yang paling mungkin dibatalkan, sehingga mengurangi biaya yang tidak perlu.

**Berdasarkan analisis sebelumnya yang mendukung High Precision sebagai pilihan yang lebih menguntungkan karena penghematan biaya, maka F0.5 score akan lebih sesuai.**

## Outliers

```python
def calculate_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

outlier_data = []

numerical = [var for var in data.columns if data[var].dtypes != 'object']

for column in numerical:
    outlier_result, lower_bound, upper_bound = calculate_outliers(data, column)
    total_outlier = len(outlier_result)
    outlier_percentage = round(total_outlier / len(data[column]) * 100, 2)
    outlier_data.append([column, total_outlier, outlier_percentage, lower_bound, upper_bound])

outlier_df = pd.DataFrame(outlier_data, columns=["Column", "Total Outliers", "Percentage (%)", "Lower Bound", "Upper Bound"])
outlier_df
```

Dari tabel ini, terlihat bahwa terdapat jumlah outlier yang signifikan pada beberapa kolom, seperti required_car_parking_spaces (16.34%) dan booking_changes (8.08%). Outlier ini menunjukkan data yang jauh berbeda dari sebagian besar data lainnya dan mungkin memerlukan penanganan khusus, seperti normalisasi atau transformasi, agar model dapat belajar secara efektif dari data tersebut.

Model machine learning akan bekerja lebih baik jika input feature dalam berada pada skala yang sama. Jika variabel numerik memiliki skala yang berbeda dapat menyebabkan model prediksi menjadi bias. **High Cardinality**

Mengurangi kardinalitas dari 162 unique value dengan mengelompokkan top 9 negara berdasarkan persentasenya pada dataset + other

```python
from sklearn.base import BaseEstimator, TransformerMixin
from typing import Literal

class ReduceCountryCardinality(BaseEstimator, TransformerMixin):
    def __init__(self, top_countries=None):
        # List of countries to keep
        if top_countries is None:
            self.top_countries = ['PRT', 'GBR', 'ESP', 'FRA', 'DEU', 'IRL', 'ITA', 'BEL', 'USA']
        else:
            self.top_countries = top_countries

    def fit(self, X, y=None):
        self.feature_names_in_ = X.columns
        return self

    def transform(self, X, y=None):
        data = X.copy()
        # Replace countries not in the list with 'Other'
        data['country'] = data['country'].apply(lambda x: x if x in self.top_countries else 'Other')
        return data
    
    def set_output(self, transform: Literal['default', 'pandas']):
        return super().set_output(transform=transform)
```

**Missing Value**

Mengisi missing value pada kolom `country` dengan Other

```python
class HandleMissingCountry(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, X, y=None):
        self.feature_names_in_ = X.columns
        return self

    def transform(self, X, y=None):
        data = X.copy()
        # Replace missing values in 'country' with 'Other'
        data['country'] = data['country'].fillna('Other')
        return data
    
    def set_output(self, transform: Literal['default', 'pandas']):
        return super().set_output(transform=transform)
```

**Polynomial Feature**

Variabel numerikal pada yaitu `required_parking_spaces`. `previous_cancellations`, `total_of_special_requests`, `booking_changes`, dan `days_in_waiting_list` akan coba diberikan transformasi berupa polynomial feature

```python
poly_columns = ['required_car_parking_spaces', 'previous_cancellations', 'total_of_special_requests','booking_changes','days_in_waiting_list']
```

```python
class PolynomialFeaturesTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, degree=2, include_bias=False, interaction_only=False, columns=None):
        self.degree = degree
        self.include_bias = include_bias
        self.interaction_only = interaction_only
        self.columns = columns  # Columns to apply polynomial features to
        self.poly = PolynomialFeatures(degree=self.degree, include_bias=self.include_bias, interaction_only=self.interaction_only)
        self.feature_names_out_ = None

    def fit(self, X, y=None):
        # Ensure columns are specified or use all columns if not specified
        if self.columns is None:
            self.columns = X.columns.tolist()
        else:
            # Check if all specified columns exist in the DataFrame
            self.columns = [col for col in self.columns if col in X.columns]
        
        # If no valid columns are found, raise an error
        if not self.columns:
            raise ValueError("No valid columns found in the input DataFrame for polynomial transformation.")
        
        # Fit polynomial features only on specified columns
        self.poly.fit(X[self.columns])
        return self

    def transform(self, X):
        # Generate polynomial features only for specified columns
        X_poly = self.poly.transform(X[self.columns])
        X_poly_df = pd.DataFrame(X_poly, columns=self.poly.get_feature_names_out(self.columns))
        
        # Drop original columns from X and reset the index
        X_remaining = X.drop(columns=self.columns).reset_index(drop=True)
        
        # Concatenate polynomial features with the rest of the data
        X_combined = pd.concat([X_remaining, X_poly_df], axis=1)
        return X_combined

    def get_feature_names_out(self, input_features=None):
        # Return feature names after transformation
        return self.poly.get_feature_names_out(self.columns)
```

**Polynomial Feature**

Penanggulangan outlier pada variabel `booking_changes`, `required_car_parking_spaces` dan `previous_cancellations`

```python
def winsorize_with_pandas(s, limits):
    return s.clip(lower=s.quantile(limits[0], interpolation='lower'), 
                  upper=s.quantile(1-limits[1], interpolation='higher'))
```

```python
from sklearn.base import OneToOneFeatureMixin
from scipy.stats.mstats import winsorize
from typing import Literal

class HandlingOutliers(BaseEstimator, TransformerMixin, OneToOneFeatureMixin):
    def __init__(self):
        pass

    def fit(self, X, y=None):
        self.feature_names_in_ = X.columns
        return self 
    
    def transform(self, X, y=None):
        data = X.copy()
        data['booking_changes'] = pd.Series(winsorize_with_pandas(data['booking_changes'], limits=(0.01, 0.01)))
        data['required_car_parking_spaces'] = pd.Series(winsorize_with_pandas(data['required_car_parking_spaces'], limits=(0.01, 0.01)))
        data['previous_cancellations'] = pd.Series(winsorize_with_pandas(data['previous_cancellations'], limits=(0.01, 0.01)))
        return data
    
    def set_output(self, transform: Literal['default', 'pandas']):
        return super().set_output(transform=transform)
```

```python
logreg = LogisticRegression(max_iter=2000, solver='lbfgs')

knn = KNeighborsClassifier()

tree = DecisionTreeClassifier(random_state=0)

logreg_pipeline = Pipeline([
    ('scaler', RobustScaler()),
    ('logistic_regression', LogisticRegression(max_iter=1000))
])

knn_pipeline = Pipeline([
    ('scaler', RobustScaler()),
    ('knn', KNeighborsClassifier())
])

dtree_pipeline = Pipeline([
    ('decision_tree', DecisionTreeClassifier(random_state=0))
])

# Meta learner for stacking
meta_logreg = LogisticRegression(max_iter=1000)

# Voting Classifier (Hard Voting)
voting_hard = VotingClassifier(
    estimators=[
        ('logreg', logreg_pipeline),
        ('knn', knn_pipeline),
        ('dtree', dtree_pipeline)
    ], voting='hard'
)

# Voting Classifier (Soft Voting)
voting_soft = VotingClassifier(
    estimators=[
        ('logreg', logreg_pipeline),
        ('knn', knn_pipeline),
        ('dtree', dtree_pipeline)
    ], voting='soft'
)

# Stacking Classifier
stacking = StackingClassifier(
    estimators=[
        ('logreg', logreg_pipeline),
        ('knn', knn_pipeline),
        ('dtree', dtree_pipeline)
    ],
    final_estimator=meta_logreg
)

# Other Models without Pipelines
rf = RandomForestClassifier(random_state=0)

adaboost = AdaBoostClassifier(random_state=0)

gradboost = GradientBoostingClassifier(random_state=0)

xgboost = XGBClassifier(random_state=0, verbosity=0)
```

Beberapa poin penting pada kode dibawah antara lain adalah:
1. Stratified K-Fold Cross-Validation:

    Metode ini membagi dataset menjadi 'k' subset atau fold (dalam hal ini bernilai 5), memastikan bahwa setiap subset memiliki proporsi dari label kelas yang sama dengan original dataset.

2. Cross-Validation:

    Cross-validation dengan menggunakan 'StratifiedKFold' digunakan untuk mengevaluasi bagaimana model-model tersebut menggeneralisasi ke dataset independen. Dataset dibagi menjadi beberapa subset training dan validasi berkali-kali dan model dilatih dan dievalusi berdasakrn subset yang berbeda-beda ini.

3. F0.5 Score:

    F0.5 score adalah rata-rata harmonis weighted dari precision dan recall, dengan lebih menekankan pada precision. Ini berguna ketika tujuannya adalah meminimalkan false positives lebih dari false negatives. Fungsi make_scorer dengan fbeta_score digunakan untuk menghitung F0.5 score untuk setiap model selama cross-validation.

4. Pipeline dan Model Evaluation:

    Kode ini membangun pipeline yang mencakup beberapa langkah preprocessing dan model itu sendiri. Pipeline kemudian dievaluasi menggunakan cross-validation, dan F0.5 score untuk setiap fold dihitung.

5. Penyimpanan Hasil Komputasi:

    Rata-rata dan standar deviasi dari F0.5 score dihitung untuk setiap model, bersama dengan skor individu dari setiap fold

```python
from sklearn.metrics import make_scorer, fbeta_score
from sklearn.model_selection import StratifiedKFold, cross_val_score
from imblearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler


# Define models and their names
fixed_models = [
    logreg_pipeline, 
    knn_pipeline, 
    dtree_pipeline, 
    gradboost, 
    adaboost, 
    voting_hard, 
    voting_soft, 
    stacking, 
    rf, 
    xgboost
]

fixed_model_names = [
    'LogisticRegression', 
    'KNeighborsClassifier', 
    'DecisionTreeClassifier', 
    'GradientBoostingClassifier', 
    'AdaBoostClassifier', 
    'VotingClassifier_Hard', 
    'VotingClassifier_Soft', 
    'StackingClassifier', 
    'RandomForestClassifier', 
    'XGBClassifier'
]

# Custom scoring function for F0.5 score
f05_scorer_fixed = make_scorer(fbeta_score, beta=0.5)

# Fixed components for the pipeline
fixed_outlier_method = HandlingOutliers()  
fixed_scaler = RobustScaler()  
fixed_resampler = None  

# Initialize lists to store results
fixed_f05_mean = []
fixed_f05_std = []
fixed_all_f05 = []
fixed_model_name = []

# Stratified K-Fold Cross-validation
fixed_skfold = StratifiedKFold(n_splits=5)

# Loop through each model to evaluate
for model, name in zip(fixed_models, fixed_model_names):
    # Create the pipeline steps with fixed preprocessing
    fixed_steps = [
        ('MissingCountry', HandleMissingCountry()), 
        ('Cardinality', ReduceCountryCardinality()),  
        ('MarketSegmentHandler', ReduceMarketSegmentCardinality()), 
        ('poly', PolynomialFeaturesTransformer(degree=1, columns=poly_columns)),
        ('outlier', fixed_outlier_method), 
        ('transformer', transformer),  
        ('scaler', fixed_scaler), 
        ('model', model)  
    ]
    
    # Build the pipeline using imblearn's Pipeline
    fixed_model_pipe = Pipeline(fixed_steps)
    
    # Cross-validation score using the custom F0.5 scorer
    fixed_model_cv = cross_val_score(
        fixed_model_pipe,
        X_train,
        y_train, 
        cv=fixed_skfold, 
        scoring=f05_scorer_fixed,
        error_score='raise'
    )

    # Store results
    fixed_f05_mean.append(fixed_model_cv.mean())
    fixed_f05_std.append(fixed_model_cv.std())
    fixed_all_f05.append(fixed_model_cv.round(4))
    fixed_model_name.append(name)

# Create a DataFrame to store the results
fixed_results_df = pd.DataFrame({
    'Model': fixed_model_name,
    'Mean F0.5': fixed_f05_mean,
    'Std F0.5': fixed_f05_std,
    'All F0.5 Scores': fixed_all_f05
})

# Sort the DataFrame by the highest Mean F0.5 score
fixed_results_df = fixed_results_df.sort_values(by='Mean F0.5', ascending=False)
```

Gradient Boosting

```python
# Define Gradient Boosting Classifier
gradient_boosting = GradientBoostingClassifier(random_state=0)

# Define hyperparameter space for Gradient Boosting Classifier including passthrough for preprocessing steps
hyperparam_space_gb = {
    'outlier': [None, HandlingOutliers(), 'passthrough'],  # Outlier handling or passthrough
    'scaler': [None, RobustScaler(), StandardScaler(), 'passthrough'],  # Scaling methods or passthrough
    'resampler': [None, SMOTE(random_state=0), RandomOverSampler(random_state=0), 'passthrough'],  # Resampling methods or passthrough
    'poly_transform': [PolynomialFeaturesTransformer(degree=d, columns=poly_columns) for d in [1, 2, 3]],  # Include polynomial degrees as hyperparameter
    'model__n_estimators': [70, 80, 90, 100, 110],  # Exploring a range around the previous best value
    'model__learning_rate': [0.08, 0.1, 0.12, 0.15],  # Finer steps around 0.1
    'model__max_depth': [2, 3, 4, 5],  # Explore depths around the best known
    'model__subsample': [0.6, 0.7, 0.8, 0.9],  # Explore subsample rates around 0.7
    'model__min_samples_split': [2, 3, 4],  # Number of samples required to split a node
    'model__min_samples_leaf': [1, 2, 3]  # Number of samples required to be at a leaf node
}

# Stratified cross-validation
skf = StratifiedKFold(n_splits=5)

# Create the pipeline for the Gradient Boosting Classifier model
estimator_gb = Pipeline([
    ('MissingCountry', HandleMissingCountry()),  # Handle missing values in 'country'
    ('Cardinality', ReduceCountryCardinality()),  # Handle unknown countries
    ('MarketSegmentHandler', ReduceMarketSegmentCardinality()),  # Handle unknown market segments
    ('outlier', 'passthrough'),  # Placeholder for outlier handling
    ('poly_transform', 'passthrough'),  # Placeholder for polynomial transformation, dynamically set by RandomizedSearchCV
    ('transformer', transformer),  # Transformer for categorical variables
    ('scaler', 'passthrough'),  # Placeholder for scaler
    ('resampler', 'passthrough'),  # Placeholder for resampling
    ('model', gradient_boosting)  # Gradient Boosting Classifier as the final model
])

# Custom scorer using F0.5 score
f05_scorer = make_scorer(fbeta_score, beta=0.5)

# Perform hyperparameter tuning using RandomizedSearchCV
tuned_gb = RandomizedSearchCV(
    estimator=estimator_gb,
    param_distributions=hyperparam_space_gb,
    cv=skf,
    scoring=f05_scorer,  # Using custom F0.5 scorer
    n_jobs=-1,
    random_state=0,
    n_iter=200  # Number of iterations for RandomizedSearchCV
)

# Fit the model to the training data
tuned_gb.fit(X_train, y_train)

# Display the best hyperparameters and the best score
print("Best Parameters:", tuned_gb.best_params_)
print("Best F0.5 Score:", tuned_gb.best_score_)
```

```python
xgboost = XGBClassifier(random_state=0, use_label_encoder=False, eval_metric='logloss')

# Refined hyperparameter space for XGBoost Classifier including preprocessing steps
hyperparam_space_xgboost = {
    'outlier': [None, HandlingOutliers(), 'passthrough'],  # Outlier handling method or passthrough
    'scaler': [None, RobustScaler(), StandardScaler(), 'passthrough'],  # Scaling methods or passthrough
    'resampler': [None, SMOTE(random_state=0), RandomOverSampler(random_state=0), 'passthrough'],  # Resampling methods or passthrough
    'poly_transform': [PolynomialFeaturesTransformer(degree=d, columns=poly_columns) for d in [1, 2, 3]],  # Include polynomial degrees as hyperparameter
    # XGBoost parameters
    'model__n_estimators': [50, 100, 150, 200],  # Number of boosting rounds
    'model__learning_rate': [0.01, 0.05, 0.1, 0.2],  # Learning rate shrinks the contribution of each tree
    'model__max_depth': [3, 4, 5, 6],  # Maximum depth of a tree
    'model__colsample_bytree': [0.6, 0.8, 1.0],  # Subsample ratio of columns when constructing each tree
    'model__subsample': [0.6, 0.8, 1.0],  # Subsample ratio of the training instance
    'model__gamma': [0, 0.1, 0.2, 0.3],  # Minimum loss reduction required to make a further partition on a leaf node
    'model__min_child_weight': [1, 3, 5]  # Minimum sum of instance weight (hessian) needed in a child
}

# Stratified cross-validation
skf = StratifiedKFold(n_splits=5)

# Create the pipeline for the XGBoost Classifier model
estimator_xgboost = Pipeline([
    ('MissingCountry', HandleMissingCountry()),  # Handle missing values in 'country'
    ('Cardinality', ReduceCountryCardinality()),  # Handle unknown countries
    ('MarketSegmentHandler', ReduceMarketSegmentCardinality()),  # Handle unknown market segments
    ('outlier', 'passthrough'),  # Placeholder for outlier handling
    ('poly_transform', 'passthrough'),  # Placeholder for polynomial transformation
    ('transformer', transformer),  # Transformer for categorical variables
    ('scaler', 'passthrough'),  # Placeholder for scaler
    ('resampler', 'passthrough'),  # Placeholder for resampling
    ('model', xgboost)  # XGBoost Classifier as the final model
])

# Custom scorer using F0.5 score
f05_scorer = make_scorer(fbeta_score, beta=0.5)

# Perform hyperparameter tuning using RandomizedSearchCV
tuned_xgboost = RandomizedSearchCV(
    estimator=estimator_xgboost,
    param_distributions=hyperparam_space_xgboost,
    cv=skf,
    scoring=f05_scorer,  # Using custom F0.5 scorer
    n_jobs=-1,
    random_state=0,
    n_iter=200  # Number of iterations for RandomizedSearchCV
)

# Fit the model to the training data
tuned_xgboost.fit(X_train, y_train)

# Display the best hyperparameters and the best score
print("Best Parameters:", tuned_xgboost.best_params_)
print("Best F0.5 Score:", tuned_xgboost.best_score_)
```

Model Stacking ini menggabungkan kekuatan beberapa algoritma dasar, yaitu Decision Tree, K-Nearest Neighbors, dan Gradient Boosting, dengan Logistic Regression sebagai meta-learner. Penggunaan beberapa model dasar ini memungkinkan Stacking Classifier untuk menangkap pola yang lebih kompleks dalam data, meningkatkan kemampuan prediksi secara keseluruhan.

**Penjelasan Stacking Classifier** 

Stacking adalah metode ensemble yang menggabungkan beberapa algoritma pembelajaran (base learners) untuk memperbaiki akurasi prediksi. Setiap base learner di Stacking Classifier menangani data dengan cara yang berbeda:
- **Decision Tree**: Membagi data ke dalam cabang keputusan berdasarkan fitur, bagus untuk menangani hubungan non-linear dalam data.
- **K-Nearest Neighbors (KNN)**: Memprediksi kelas berdasarkan jarak dari contoh-contoh yang telah dilabeli sebelumnya, baik untuk data dengan pola lokal.
- **Gradient Boosting**: Model boosting yang kuat yang membangun model dari prediksi yang salah dari model sebelumnya, efektif dalam menangani data yang kompleks dan menangkap interaksi antar fitur.

Kemudian, Logistic Regression digunakan sebagai meta-learner, mengambil prediksi dari base learners sebagai input dan menghasilkan prediksi akhir. Ini memungkinkan model untuk "belajar" dari kesalahan base learners dan meningkatkan akurasi prediksi keseluruhan.

**Metrik yang Digunakan untuk Evaluasi**

Dalam evaluasi model, **F0.5 score** digunakan sebagai metrik utama. F0.5 score adalah versi dari F-score yang lebih menekankan pada precision dibandingkan recall. Ini berarti bahwa model lebih dihargai karena memprediksi positif secara akurat (mengurangi false positives), yang sangat penting dalam kasus prediksi pembatalan reservasi hotel, di mana salah memprediksi bahwa tamu akan datang tetapi sebenarnya mereka membatalkan dapat memiliki dampak operasional dan finansial yang signifikan.

## E.11  Actual vs Predicted Probability

Calibration Plot membandingkan probabilitas prediksi model dengan probabilitas sebenarnya dari hasil yang diamati. Tujuan dari pembuatan plot ini adalah membantu memahami seberapa baik probabilitas prediksi mencerminkan kemungkinan sebenarnya dari suatu kejadian yaitu untuk menilai probabilitas model memprediksi pembatalan booking hotel.

```python
y_proba = tuned_model.predict_proba(X_test)[:,1]
```

```python
if isinstance(y_test, pd.Series):
    df_proba = pd.DataFrame({'Exited': y_test, 'probability': y_proba})
elif isinstance(y_test, pd.DataFrame):
    df_proba = y_test.copy()
    df_proba['probability'] = y_proba
else:
    raise ValueError("y_test should be a pandas Series or DataFrame.")
```

```python
def probability_range(x):
    if x < 0.1:
        return '0.0 - 0.1'
    elif 0.1 <= x < 0.2:
        return '0.1 - 0.2'
    elif 0.2 <= x < 0.3:
        return '0.2 - 0.3'
    elif 0.3 <= x < 0.4:
        return '0.3 - 0.4'
    elif 0.4 <= x < 0.5:
        return '0.4 - 0.5'
    elif 0.5 <= x < 0.6:
        return '0.5 - 0.6'
    elif 0.6 <= x < 0.7:
        return '0.6 - 0.7'
    elif 0.7 <= x < 0.8:
        return '0.7 - 0.8'
    elif 0.8 <= x < 0.9:
        return '0.8 - 0.9'
    else:
        return '0.9 - 1.0'

df_proba['probability range'] = df_proba['probability'].apply(probability_range)

df_probability = df_proba.groupby(['probability range', 'Exited']).size().unstack(fill_value=0).reset_index()

df_probability['total'] = df_probability[0] + df_probability[1]

df_probability['actual probability'] = (df_probability[1] / df_probability['total']).round(2)

df_probability['predicted probability'] = df_proba.groupby(['probability range'])['probability'].mean().values.round(2)

df_probability = df_probability.sort_values(by='probability range', ascending=False)

df_probability = df_probability.rename(columns={0: 'no', 1: 'yes'})

df_probability
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(data=df_probability.sort_values(by='probability range', ascending=True), x='predicted probability', y='actual probability')
value = [x/100 for x in range(0,110,10)]
sns.lineplot(x=value, y=value)
plt.xlim(0,1)
plt.ylim(0,1)
plt.axvline(0.5, color='r', linestyle='--')
plt.annotate('Threshold = 0.5', (0.5,0.95))
plt.title('Calibration Plot')
plt.show()
```

1. Calibration Line (Garis Orange):
- Garis diagonal orange menunjukkan model yang terkalibrasi dengan sempurna, dimana probabilitas yang diprediksi cocok dengan probabilitas yang sebenarnya.

2. Model Calibration (Garis Biru):
- Garis biru menunjukkan kalibrasi sebenarnya dari model yang menunjukkan hubungan antara probabilitas yang diprediksi dan frekuensi pembatalan yang diamati pada data test.
- Pada skenario ideal maka garis biru harus mengikuti garis orange yang menunjukkan bahwa probabilitas yang diprediksi model terkalibrasi dengan baik dan secara akurat merefleksikan kejadian yang sebenarnya.

3. Interpretation:
- Garis biru secara umum mengikuti pola dari garis diagonal orange yang menunjukkan bahwa prediksi model dapat diandalkan dan mendekati kemungkinan pembatalan.
- Kedekatan garis biru dengan garis oranye di seluruh rentang probabilitas (dari 0 hingga 1) berarti bahwa baik ketika model memprediksi probabilitas pembatalan rendah, sedang, maupun tinggi, prediksi-prediksi ini konsisten dengan hasil yang sebenarnya.
